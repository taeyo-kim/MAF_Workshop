## Step 3.2: Function Tools ì •ì˜

````python
from typing import Annotated
from pydantic import Field
from agent_framework import tool

@tool
def get_flight_prices(
    origin: Annotated[str, Field(description="ì¶œë°œ ë„ì‹œ ğŸ¬")],
    destination: Annotated[str, Field(description="ë„ì°© ë„ì‹œ ğŸ¯")]
) -> str:
    """ë‘ ë„ì‹œ ê°„ í•­ê³µí¸ ì •ë³´ë¥¼ ì¡°íšŒí•©ë‹ˆë‹¤. âœˆï¸ğŸ’°"""
    route = f"{origin}-{destination}"
    if route in MOCK_FLIGHTS:
        f = MOCK_FLIGHTS[route]
        return f"{origin}â†’{destination}: ${f['price']}, {f['airline']}, {f['duration']}"
    return f"{origin}â†’{destination} ì§í•­í¸ ì—†ìŒ"

@tool
def search_hotels(
    city: Annotated[str, Field(description="ë„ì‹œ ì´ë¦„ ğŸŒ†")]
) -> str:
    """ë„ì‹œì˜ í˜¸í…” ëª©ë¡ì„ ê²€ìƒ‰í•©ë‹ˆë‹¤. ğŸ¨âœ¨"""
    if city in MOCK_HOTELS:
        result = f"{city} í˜¸í…”:\n"
        for h in MOCK_HOTELS[city]:
            result += f"- {h['name']}: ${h['price']}/ë°•, â­{h['rating']}\n"
        return result
    return f"{city} í˜¸í…” ì •ë³´ ì—†ìŒ"

@tool
def get_weather(
    city: Annotated[str, Field(description="ë„ì‹œ ì´ë¦„ ğŸŒ")]
) -> str:
    """ë„ì‹œì˜ í˜„ì¬ ë‚ ì”¨ë¥¼ ì¡°íšŒí•©ë‹ˆë‹¤. â˜€ï¸ğŸŒ§ï¸"""
    if city in MOCK_WEATHER:
        w = MOCK_WEATHER[city]
        return f"{city}: {w['temp']}, {w['condition']} ({w['season']})"
    return f"{city} ë‚ ì”¨ ì •ë³´ ì—†ìŒ"

print("âœ… Function Tools ì •ì˜ ì™„ë£Œ")
````

- `invoked()` : LLM ì‘ë‹µ ì´í›„ ì‹¤í–‰. ì‚¬ìš©ì ë©”ì‹œì§€ì—ì„œ ì„ í˜¸ë„ ë“±ì„ ì¶”ì¶œí•´ ë‚´ë¶€ ìƒíƒœ ì—…ë°ì´íŠ¸

## Step 5.1: ë©”ëª¨ë¦¬ í´ë˜ìŠ¤ ì •ì˜

````python
from agent_framework import ContextProvider, Context, ChatMessage
from collections.abc import Sequence
from typing import Any

class TravelPreferencesMemory(ContextProvider):
    """ì‚¬ìš©ì ì—¬í–‰ ì„ í˜¸ë„ë¥¼ ê¸°ì–µí•˜ëŠ” ë©”ëª¨ë¦¬ ğŸ§ âœ¨"""
    
    def __init__(self):
        self.preferences = {
            "budget": None,
            "style": None,
            "interests": []
        }
    
    def extract_preferences(self, text: str) -> None:
        """í…ìŠ¤íŠ¸ì—ì„œ ì„ í˜¸ë„ ì¶”ì¶œ (ì§ì ‘ í˜¸ì¶œìš©) ğŸ”âœ¨"""
        text = text.lower()
        
        # ì˜ˆì‚°
        if "ì €ë ´" in text or "ê°€ì„±ë¹„" in text:
            self.preferences["budget"] = "ì €ë ´/ê°€ì„±ë¹„"
        elif "ëŸ­ì…”ë¦¬" in text or "ê³ ê¸‰" in text:
            self.preferences["budget"] = "ëŸ­ì…”ë¦¬"
        
        # ìŠ¤íƒ€ì¼
        if "ëª¨í—˜" in text or "ì•¡í‹°ë¹„í‹°" in text:
            self.preferences["style"] = "ëª¨í—˜/ì•¡í‹°ë¹„í‹°"
        elif "íœ´ì–‘" in text or "íœ´ì‹" in text:
            self.preferences["style"] = "íœ´ì–‘/íœ´ì‹"
        elif "ë¬¸í™”" in text or "ì—­ì‚¬" in text:
            self.preferences["style"] = "ë¬¸í™”/ì—­ì‚¬"
        
        # ê´€ì‹¬ì‚¬
        for interest, keywords in [("ìŒì‹", ["ìŒì‹", "ë§›ì§‘"]), 
                                   ("ì‡¼í•‘", ["ì‡¼í•‘"]), 
                                   ("ìì—°", ["ìì—°", "ì‚°", "ë°”ë‹¤"])]:
            if any(k in text for k in keywords) and interest not in self.preferences["interests"]:
                self.preferences["interests"].append(interest)
    
    async def invoking(self, messages=None, **kwargs) -> Context | None:
        """ì—ì´ì „íŠ¸ í˜¸ì¶œ ì „: ë©”ëª¨ë¦¬ ì œê³µ ğŸ“¤âœ¨"""
        if any(self.preferences.values()):
            info = f"""
[ì‚¬ìš©ì ì„ í˜¸ë„ ì •ë³´]
- ì˜ˆì‚°: {self.preferences['budget'] or 'ë¯¸ì •'}
- ì—¬í–‰ ìŠ¤íƒ€ì¼: {self.preferences['style'] or 'ë¯¸ì •'}
- ê´€ì‹¬ì‚¬: {', '.join(self.preferences['interests']) or 'ë¯¸ì •'}

ìœ„ ì •ë³´ë¥¼ ì°¸ê³ í•˜ì—¬ ê°œì¸í™”ëœ ì¶”ì²œì„ ì œê³µí•˜ì„¸ìš”.
"""
            return Context(instructions=info)
        return None
    
    async def invoked(
        self,
        request_messages: ChatMessage | Sequence[ChatMessage],
        response_messages: ChatMessage | Sequence[ChatMessage] | None = None,
        invoke_exception: Exception | None = None,
        **kwargs: Any,
    ) -> None:
        """ì—ì´ì „íŠ¸ í˜¸ì¶œ í›„: ì„ í˜¸ë„ ì¶”ì¶œ (ìë™)"""
        # request_messagesë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
        if isinstance(request_messages, ChatMessage):
            messages_list = [request_messages]
        else:
            messages_list = list(request_messages)
        
        # user ë©”ì‹œì§€ì—ì„œ í…ìŠ¤íŠ¸ ì¶”ì¶œ
        for msg in messages_list:
            if hasattr(msg, 'role') and msg.role.value == "user":
                text = ""
                if hasattr(msg, 'contents') and msg.contents:
                    for content in msg.contents:
                        if hasattr(content, 'text'):
                            text += content.text
                        elif isinstance(content, str):
                            text += content
                        elif isinstance(content, dict) and 'text' in content:
                            text += content['text']
                
                if text:
                    self.extract_preferences(text)

print("âœ… TravelPreferencesMemory ì •ì˜ ì™„ë£Œ")
````

## Step 5.2: ë©”ëª¨ë¦¬ ì‚¬ìš© í…ŒìŠ¤íŠ¸

````python
from agent_framework import ChatAgent

# ë©”ëª¨ë¦¬ ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
memory = TravelPreferencesMemory()

# ChatClient ê°€ì ¸ì˜¤ê¸°
chat_client = AzureOpenAIChatClient(credential=AzureCliCredential())

# ë©”ëª¨ë¦¬ê°€ ìˆëŠ” ì—ì´ì „íŠ¸ ìƒì„± (context_provider ë‹¨ìˆ˜!)
agent_with_memory = ChatAgent(
    chat_client=chat_client,
    instructions="""
    ë‹¹ì‹ ì€ ì—¬í–‰ ê³„íš ì–´ì‹œìŠ¤í„´íŠ¸ì…ë‹ˆë‹¤. âœˆï¸âœ¨
    ì‚¬ìš©ìì˜ ì„ í˜¸ë„ë¥¼ ê¸°ì–µí•˜ê³  ê°œì¸í™”ëœ ì¶”ì²œì„ ì œê³µí•©ë‹ˆë‹¤. ğŸ§ ğŸ’–
    ë„êµ¬ë¥¼ í™œìš©í•˜ì—¬ ì •í™•í•œ ì •ë³´ë¥¼ ì œê³µí•˜ì„¸ìš”. ğŸ”§
    í•œêµ­ì–´ë¡œ ì‘ë‹µí•©ë‹ˆë‹¤. ğŸ‘‹
    """,
    tools=[get_flight_prices, search_hotels, get_weather],
    context_provider=memory  # í•µì‹¬! ë‹¨ìˆ˜ë¡œ ì „ë‹¬
)

print("âœ… ë©”ëª¨ë¦¬ê°€ ìˆëŠ” ì—ì´ì „íŠ¸ ìƒì„± ì™„ë£Œ")
````

````python
from agent_framework import WorkflowBuilder, WorkflowContext, WorkflowOutputEvent, executor
from typing_extensions import Never

@executor(id="parse_request")
async def parse_request(text: str, ctx: WorkflowContext[dict]) -> None:
    """Step 1: ìš”ì²­ íŒŒì‹± ğŸ”âœ¨"""
    print("ğŸ” Step 1: ìš”ì²­ íŒŒì‹± ì¤‘...")
    # ê°„ë‹¨í•œ íŒŒì‹± ë¡œì§
    data = {"destination": "ë„ì¿„", "duration": "3ì¼", "interests": ["ë¬¸í™”", "ìŒì‹"]}
    await ctx.send_message(data)

@executor(id="generate_itinerary")
async def generate_itinerary(data: dict, ctx: WorkflowContext[Never, str]) -> None:
    """Step 2: ì¼ì • ìƒì„± ğŸ“…âœ¨"""
    print("ğŸ“ Step 2: ì¼ì • ìƒì„± ì¤‘...")
    itinerary = f"""
ğŸ—¾ {data['destination']} {data['duration']} ì¼ì •

ğŸ“… Day 1: {data['interests'][0]} íƒë°©
ğŸ“… Day 2: {data['interests'][1]} íˆ¬ì–´  
ğŸ“… Day 3: ììœ ì‹œê°„ & ì¶œë°œ
"""
    await ctx.yield_output(itinerary)

print("âœ… ì›Œí¬í”Œë¡œìš° ë‹¨ê³„ ì •ì˜ ì™„ë£Œ")
````

## Step 6.2: ì›Œí¬í”Œë¡œìš° ì‹¤í–‰

````python
# ì›Œí¬í”Œë¡œìš° ë¹Œë“œ (add_edge + set_start_executor)
workflow = (
    WorkflowBuilder()
    .add_edge(parse_request, generate_itinerary)  # ì—°ê²°
    .set_start_executor(parse_request)             # ì‹œì‘ì 
    .build()
)

print("ğŸš€ Workflow ì‹¤í–‰\n")

async for event in workflow.run_stream("ë„ì¿„ 3ì¼ ë¬¸í™”/ìŒì‹ ì—¬í–‰ ğŸ‡ªğŸ‡µğŸœ"):
    print(f"ğŸ“Œ ì´ë²¤íŠ¸: {type(event).__name__}")
    if isinstance(event, WorkflowOutputEvent):
        print("\nğŸ“Š ìµœì¢… ê²°ê³¼:")
        print(event.data)
````

## Step 6.3: ì—ì´ì „íŠ¸ ê¸°ë°˜ ì›Œí¬í”Œë¡œìš° (Writer â†’ Reviewer)

ì´ë²ˆì—” **AI ì—ì´ì „íŠ¸**ë¥¼ ì›Œí¬í”Œë¡œìš° ë…¸ë“œë¡œ ì‚¬ìš©í•©ë‹ˆë‹¤:
- `register_agent()` - ì—ì´ì „íŠ¸ë¥¼ ì›Œí¬í”Œë¡œìš° ë…¸ë“œë¡œ ë“±ë¡
- Writer ì—ì´ì „íŠ¸ â†’ Reviewer ì—ì´ì „íŠ¸ ìˆœì°¨ ì‹¤í–‰

````python
from agent_framework import AgentRunUpdateEvent

# ChatClient ìƒì„± (async with ë¶ˆí•„ìš”)
client = AzureOpenAIChatClient(credential=AzureCliCredential())

# ì—ì´ì „íŠ¸ ê¸°ë°˜ ì›Œí¬í”Œë¡œìš° ë¹Œë“œ
agent_workflow = (
    WorkflowBuilder()
    # Writer ì—ì´ì „íŠ¸ ë“±ë¡
    .register_agent(
        lambda: client.as_agent(
            name="TravelWriter",
            instructions="ë‹¹ì‹ ì€ ì—¬í–‰ ì½˜í…ì¸  ì‘ê°€ì…ë‹ˆë‹¤. ì£¼ì–´ì§„ ì£¼ì œë¡œ ì§§ê³  ë§¤ë ¥ì ì¸ ì—¬í–‰ ì†Œê°œê¸€(3-4ë¬¸ì¥)ì„ ì‘ì„±í•˜ì„¸ìš”. í•œêµ­ì–´ë¡œ ì‘ì„±í•©ë‹ˆë‹¤. âœï¸âœ¨"
        ),
        name="writer"
    )
    # Reviewer ì—ì´ì „íŠ¸ ë“±ë¡ (output_response=Trueë¡œ ìµœì¢… ì¶œë ¥)
    .register_agent(
        lambda: client.as_agent(
            name="TravelReviewer", 
            instructions="ë‹¹ì‹ ì€ ì—¬í–‰ ì½˜í…ì¸  í¸ì§‘ìì…ë‹ˆë‹¤. ì‘ì„±ëœ ê¸€ì„ ê²€í† í•˜ê³  ê°„ë‹¨í•œ í”¼ë“œë°±(2-3ë¬¸ì¥)ì„ ì œê³µí•˜ì„¸ìš”. í•œêµ­ì–´ë¡œ í”¼ë“œë°±í•©ë‹ˆë‹¤. ğŸ“âœ¨"
        ),
        name="reviewer",
        output_response=True  # ì´ ì—ì´ì „íŠ¸ì˜ ì‘ë‹µì´ ìµœì¢… ì¶œë ¥
    )
    .set_start_executor("writer")
    .add_edge("writer", "reviewer")
    .build()
)

print("ğŸš€ ì—ì´ì „íŠ¸ ê¸°ë°˜ ì›Œí¬í”Œë¡œìš° ì‹¤í–‰\n")
print("=" * 50)

last_executor = None
async for event in agent_workflow.run_stream("ë„ì¿„ ë²¤ê½ƒ ì—¬í–‰ ğŸŒ¸ğŸ‡ªğŸ‡µ"):
    if isinstance(event, AgentRunUpdateEvent):
        # ì—ì´ì „íŠ¸ê°€ ë°”ë€Œë©´ ì¤„ë°”ê¿ˆ
        if event.executor_id != last_executor:
            if last_executor:
                print("\n" + "=" * 50)
            print(f"ğŸ¤– [{event.executor_id}]:", end=" ")
            last_executor = event.executor_id
        print(event.data, end="", flush=True)
    elif isinstance(event, WorkflowOutputEvent):
        print("\n\nâœ… ì›Œí¬í”Œë¡œìš° ì™„ë£Œ! ğŸ‰")
````
