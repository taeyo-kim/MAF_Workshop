# ⚡ CopilotKit으로 AI 에이전트 UI 빠르게 구성하기

---

## 📌 CopilotKit이란?

**CopilotKit**은 AI 기반 사용자 대상 에이전트 애플리케이션을 구축하기 위한 **오픈 소스 에이전트 애플리케이션 플랫폼**입니다.

> 애플리케이션의 **논리 · 상태 · UI · 컨텍스트**를 에이전트 백엔드에 연결하여,  
> 임베디드 UI와 헤드리스 인터페이스 전반에 걸쳐 상호작용적인 경험을 가능하게 합니다.

---

## 🔌 주요 통합 지원

| 연결 방식 | 설명 |
|-----------|------|
| **직접 연결** | 에이전트 백엔드에 직접 연결 |
| **AG-UI 프로토콜** | Microsoft Agent Framework 포함 주요 프레임워크 지원 |
| **MCP / A2A** | 다양한 에이전트 프로토콜 지원 |

---

## ✨ 핵심 장점

- 🔄 **프레임워크 독립성** — 특정 프레임워크·모델·에이전트 프로토콜에 종속되지 않음
- 🚀 **빠른 UI 제공** — 앱의 UX를 재설계하지 않고도 AI 스택을 자유롭게 진화
- 🤝 **MAF 통합** — Microsoft Agent Framework와 AG-UI 프로토콜을 통해 원활하게 연동

### ✅ 사전 준비

> **⚠️ 주의:**   
> 다음 단계를 진행하기 전에 이전 과정에서 만든 `agent_server.py` 서버가 실행 중이어야 합니다.   
> 서버를 종료했다면, 기존 터미널에서 아래 명령으로 재시작하세요. (폴더는 .\AG-UI 기준)  
> 터미널을 닫았다면, 새로운 터미널을 열고 venv 활성화 후 아래 명령으로 서버를 실행하세요.  
>
> ```powershell
> .\.venv\Scripts\Activate.ps1
> cd .\AG-UI\
> python agent_server.py
> ```
