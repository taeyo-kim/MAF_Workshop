# Microsoft Agent Framework

MAF(Microsoft Agent Framework)는 .NET 및 Python 환경에서 AI 에이전트 와 다중 에이전트 워크플로우를 구축하기 위한 오픈소스 개발 키트입니다. Semantic Kernel 과 AutoGen 프로젝트의 아이디어를 통합하고 확장하여 두 프로젝트의 강점을 결합하고 새로운 기능을 추가했습니다. 동일한 개발팀에서 개발하고 있으므로, 이 프레임워크는 향후 AI 에이전트 구축을 위한 통합 기반을 제공할 것입니다.

에이전트 프레임워크는 크게 두 가지 범주의 기능을 제공합니다.

- AI 에이전트(AI Agents) : LLM을 사용하여 사용자 입력을 처리하고, 도구 및 MCP 서버를 호출하여 작업을 수행하고, 응답을 생성하는 개별 에이전트입니다. 에이전트는 Azure OpenAI, OpenAI 및 Azure AI를 포함한 모델 공급자를 지원합니다.

- 워크플로우(Workflows) : 여러 에이전트와 기능을 연결하여 복잡하고 다단계적인 작업을 수행하는 그래프 기반 워크플로우입니다. 워크플로우는 형식-기반(type-based) 라우팅, 중첩, 체크포인트를 지원하고, 사람이 개입하는 시나리오(human-in-the-loop)를 위한 요청/응답 패턴도 지원합니다.

MAF 프레임워크는 모델 클라이언트(chat completion 및 response), 상태 관리를 위한 에이전트 스레드, 에이전트 메모리를 위한 컨텍스트 제공자, 에이전트 동작을 가로채는 미들웨어, 그리고 도구 통합을 위한 MCP 클라이언트 등 기본적인 빌딩 블록 요소들을 제공합니다. 이러한 구성요소들을 통해 상호작용적이고, 견고하며, 안전한 AI 애플리케이션을 구축할 수 있는 유연성과 강력한 기능을 제공합니다.

### AI Agents 란

AI 에이전트는 LLM을 사용하여 사용자 입력을 처리하고, 의사 결정을 내리고, 도구 및 MCP 서버를 호출하여 작업을 수행하고, 응답을 생성합니다. 다음의 다이어그램은 AI 에이전트의 핵심 구성 요소와 그 상호 작용을 보여줍니다.

![diagram-agent](images/agent-diagram.png)

AI 에이전트는 또한 스레드, 컨텍스트 제공자 및 미들웨어와 같은 추가 구성요소를 사용하여 그 기능을 향상시킬 수 있습니다. 더 자세한 내용은 [공식 사이트의 문서](https://learn.microsoft.com/en-us/agent-framework/overview/agent-framework-overview)를 참고하시기 바랍니다.

### Workflow란

워크플로우는 AI 에이전트를 구성요소로 포함할 수 있는 미리-정의된 작업들의 시퀀스를 표현할 수 있으며, 일관성과 신뢰성을 유지합니다. 워크플로우는 멀티 에이전트, 사람과의 상호작용, 외부 시스템과의 통합을 포함하는 복잡하고 장기-실행 프로세스를 처리하도록 설계되었습니다.

워크플로우의 실행 시퀀스는 명시적으로 정의될 수 있으며, 실행 경로에 대한 더 많은 제어를 허용합니다. 다음의 다이어그램은 두 개의 AI 에이전트와 하나의 함수를 연결하는 워크플로우의 예시를 보여줍니다.

![diagram-wf](images/wf-diagram.png)

워크플로우는 또한 조건부 라우팅, 모델 기반 의사 결정, 동시 실행을 사용하여 동적 시퀀스를 표현할 수 있습니다. 이것이 멀티-에이전트 오케스트레이션 패턴이 구현되는 방식입니다. 오케스트레이션 패턴은 여러 에이전트가 복잡한 작업을 수행하도록 조정하는 메커니즘을 제공하며, 이는 단일 에이전트의 한계를 극복합니다. 더 자세한 내용은 [공식 사이트의 문서](https://learn.microsoft.com/en-us/agent-framework/overview/agent-framework-overview)를 참고하시기 바랍니다.
