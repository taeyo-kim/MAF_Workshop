## 🔌 AzureAIAgentClient에 MCP 도구 연결하기

이번 실습에서는 Foundry에서 호스팅하는 **Microsoft Learn MCP 서버**를 사용하여, 에이전트에 MCP 도구를 연결하는 방법을 다룹니다. `client.get_mcp_tool()` 메서드를 활용하면 Hosted MCP 서버를 손쉽게 에이전트에 통합할 수 있습니다.

<div style="background: #EFF6FC; border-left: 4px solid #0078D4; border-radius: 0 8px 8px 0; padding: 14px 20px; margin: 16px 0; color: #323130;max-width: 800px;">
  <div style="font-weight: 600; color: #0078D4; margin-bottom: 6px;">ℹ️ Microsoft Learn MCP 서버란?</div>
  <p style="margin: 0; line-height: 1.7;">Microsoft Learn MCP는 별도의 인증이나 API 키 없이 누구나 사용할 수 있는 <strong>공개 MCP 서버</strong>입니다. 이를 통해 Microsoft 공식 기술 문서를 검색하는 에이전트를 손쉽게 구성할 수 있습니다.</p>
</div>

아래 코드는 `AzureAIAgentClient`로 Microsoft Learn MCP 도구를 연결하고, 에이전트를 실행하는 기본 예제입니다.

</div>


````python
# 코드 예시 ...

async def basic_foundry_mcp_example():
    """Hosted MCP 도구를 사용하는 Foundry 에이전트의 기본 예시"""
    
    async with (
        AzureCliCredential() as credential,
        AzureAIAgentClient(credential=credential) as client,
    ):

        # client.get_mcp_tool() 메서드를 사용하여
        # Foundry에서 호스팅되는 MCP 도구를 가져올 수 있습니다.
        learn_mcp = client.get_mcp_tool(
            name="Microsoft Learn MCP",
            url="https://learn.microsoft.com/api/mcp",
            approval_mode="never",  # 승인 없이 MCP 도구 자동 실행
        )

        # Hosted MCP 도구를 사용하는 에이전트를 생성합니다.
        agent = client.as_agent(
            name="MicrosoftLearnAgent", 
            instructions="당신은 Microsoft Learn 관련 내용은 MCP를 활용해서 질문에 답변합니다."
                        +"MCP를 사용할 수 없는 질문은 모른다고 답변해야 합니다.",
            tools=learn_mcp,
        )

        session = agent.create_session()
        result = await agent.run(
            "롱블랙은 어떻게 만드나요?",
            session=session,
        )
        print(result.text)

if __name__ == "__main__":
    asyncio.run(basic_foundry_mcp_example())

# 현재 폴더에는 완성된 agent_mcp_hosted.py 파일이 제공됩니다. 
# 실행을 위해서는 우선 터미널을 열고, 가상환경을 활성화 한 후에 다음 명령어를 실행합니다.
#
# .\.venv\Scripts\Activate.ps1; 
# python .\MAF\agent_mcp_hosted.py
````

현재 폴더에는 완성된 agent_mcp_hosted.py 파일이 제공됩니다.   
실행을 위해서는 우선 터미널을 열고, 가상환경을 활성화 한 후에 다음 명령어를 실행합니다.

이번에는 에이전트가 **Microsoft Learn MCP 도구를 활용**하여 기술 문서를 검색하는 것을 확인할 수 있습니다.

<div style="background: #EFF6FC; border-left: 4px solid #0078D4; border-radius: 0 8px 8px 0; padding: 14px 20px; margin: 16px 0; color: #323130;max-width: 800px;">
  <div style="font-weight: 600; color: #0078D4; margin-bottom: 6px;">💡 MCP 호출 내역 확인</div>
  <p style="margin: 0; line-height: 1.7;">응답 결과만으로는 에이전트가 실제로 MCP 서버의 도구를 호출했는지 확인하기 어렵습니다. 코드를 다음과 같이 변경하여 <strong>MCP 호출 내역(Step Details)</strong>을 명확히 검증해 봅니다.</p>
</div>

</div>


````python
# 코드 예시 ... 상위 코드는 기존과 동일합니다

        session = agent.create_session()
        result = await agent.run(
            "Azure AI Agent 문서에서 MCP 도구 호출과 관련된 내용을 요약해 주세요.",
            session=session,
        )

        # MCP 호출 이력 출력
        print("=" * 60)
        print("📋 MCP 호출 이력")
        print("=" * 60)
        thread_id = session.service_session_id
        agents_client = client.agents_client
        mcp_call_count = 0
        runs = agents_client.runs.list(thread_id=thread_id)
        async for run in runs:
            steps = agents_client.run_steps.list(thread_id=thread_id, run_id=run.id)
            async for step in steps:
                if step.type == "tool_calls" and hasattr(step, "step_details"):
                    for tc in step.step_details.tool_calls:
                        if tc.type == "mcp":
                            mcp_call_count += 1
                            print(f"\n[호출 #{mcp_call_count}]")
                            print(f"  서버  : {tc.server_label}")
                            print(f"  도구  : {tc.name}")
                            print(f"  인수  : {tc.arguments}")
                            print(f"  결과  : {str(tc.output)[:300]}...")
        if mcp_call_count == 0:
            print("  (MCP 호출 없음)")
        print("=" * 60)
        print("\n💬 최종 응답:")
        print(result.text)


if __name__ == "__main__":
    asyncio.run(basic_foundry_mcp_example())

# 현재 폴더에는 완성된 agent_mcp_step_details.py 파일이 제공됩니다. 
# 실행을 위해서는 우선 터미널을 열고, 가상환경을 활성화 한 후에 다음 명령어를 실행합니다.
#
# .\.venv\Scripts\Activate.ps1; 
# python .\MAF\agent_mcp_step_details.py
````

## 🧩 멀티 도구 등록하기

`AzureAIAgentClient`에는 **여러 개의 MCP 도구를 동시에 등록**할 수 있습니다. 예를 들어, Microsoft Learn MCP 도구와 함께 **GitHub MCP 도구**를 함께 등록하면, 에이전트가 두 서비스 모두를 활용하여 답변을 생성할 수 있습니다.

<div style="background: #EFF6FC; border-left: 4px solid #0078D4; border-radius: 0 8px 8px 0; padding: 14px 20px; margin: 16px 0; color: #323130;max-width: 800px;">
  <div style="font-weight: 600; color: #0078D4; margin-bottom: 6px;">🔑 GitHub MCP 사용 시 주의사항</div>
  <p style="margin: 0; line-height: 1.7;">GitHub MCP 도구를 사용하려면 GitHub 계정의 <strong>Personal Access Token(PAT)</strong>이 필요합니다. PAT는 GitHub 설정 페이지에서 발급받을 수 있으며, 코드의 <code>Authorization</code> 헤더에 입력해야 합니다.</p>
</div>

아래 코드는 `get_mcp_tool()` 메서드를 두 번 호출하여 **Microsoft Learn**과 **GitHub**를 동시에 에이전트 도구로 등록하는 예제입니다.

</div>


````python
# 상기 코드는 동일 ...

    # get_mcp_tool() 메서드를 사용하여 여러 MCP 도구를 생성할 수 있습니다.
    # Microsoft Learn MCP 도구 생성
    learn_mcp = client.get_mcp_tool(
        name="Microsoft Learn MCP",
        url="https://learn.microsoft.com/api/mcp",
        approval_mode="never",  # 승인 없이 MCP 도구 자동 실행
    )
    # GitHub MCP 도구 생성
    github_mcp = client.get_mcp_tool(
        name="GitHub MCP", 
        url="https://api.githubcopilot.com/mcp",
        approval_mode="never"   # 승인이 필요하다면 "always_require" 설정
        headers={"Authorization": "Bearer <GITHUB 계정의 PAT를 입력하세요>"},
    )

    # 위에서 생성한 여러 MCP 도구를 등록할 수 있습니다.
    agent = client.as_agent(
        name="MultiToolAgent",
        instructions="당신은 MS Learn 기술문서와 GitHub 저장소를 검색하여 정보를 제공합니다."
                    +"MCP를 사용할 수 없는 질문은 모른다고 답변해야 합니다.",
        tools=[learn_mcp, github_mcp],
    )

    # 질문
    result = await agent.run(
                "내 깃헙의 리포들을 나열해 주세요", 
                session=session)

# 이하 코드 동일 ...

# 현재 폴더에는 완성된 agent_mcp_hosted2.py 파일이 제공됩니다. 
# 실행을 위해서는 우선 터미널을 열고, 가상환경을 활성화 한 후에 다음 명령어를 실행합니다.
#
# .\.venv\Scripts\Activate.ps1; 
# python .\MAF\agent_mcp_hosted2.py
````

현재 폴더에는 완성된 agent_mcp_hosted2.py 파일이 제공됩니다.   
실행을 위해서는 우선 터미널을 열고, 가상환경을 활성화 한 후에 다음 명령어를 실행합니다.

``` powershell
    .\.venv\Scripts\Activate.ps1; 
    python .\MAF\agent_mcp_hosted2.py
```
