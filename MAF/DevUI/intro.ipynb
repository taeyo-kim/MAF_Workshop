# 🎨 DevUI - 에이전트 및 워크플로우 실행을 위한 앱

DevUI는 Microsoft Agent Framework를 사용하는 에이전트와 워크플로우를 실행 및 테스트하기 위한 경량의 독립형 샘플 애플리케이션입니다.

### 🚀 주요 기능

DevUI는 대화형 테스트를 위한 웹 인터페이스와 OpenAI 호환 API 백엔드를 제공하여, 구축한 에이전트와 워크플로우를 여러분의 애플리케이션에 통합하기 전에 시각적으로 디버깅, 테스트 및 반복 작업할 수 있도록 지원합니다.

> **⚠️ 중요**  
>  
> DevUI는 개발 중에 에이전트와 워크플로우를 시각화하고 디버깅하는 데 도움이 되는 샘플 앱입니다.  
> 운영(Production) 환경에서 사용하기 위한 것은 아닙니다.

## 📂 샘플 프로젝트 구성

현재 DevUI 폴더에는 이미 2개의 샘플(에이전트 하나와 워크플로우 하나)이 포함되어 있습니다.   
각각의 샘플은 개별 디렉토리 안에 Python 코드로 작성되어 있습니다.

### 🤖 포함된 샘플

#### 1️⃣ Weather Agent
📁 **경로**: `weather_agent/agent.py`
- 간단한 날씨 정보 제공 에이전트 샘플입니다
- 사용자가 입력한 위치에 대한 현재 날씨 정보를 제공합니다
- 앞서 실습에서 사용했던 날씨 에이전트를 온전한 Python 코드로 작성한 것입니다

#### 2️⃣ Multi-Agent Workflow
📁 **경로**: `workflow_ex/workflow.py`
- 다중 에이전트 워크플로우 샘플입니다
- 작성자(Writer), 검토자(Reviewer), 편집자(Editor), 출판자(Publisher)로 구성된 에이전트들이 협력합니다
- 콘텐츠를 생성하고 검토하는 과정을 시뮬레이션합니다
- 품질 점수에 따라 검토 경로가 달라집니다

## 🚀 DevUI 실행하기

DevUI 도구는 디렉터리 기반 검색을 지원하기에, 에이전트 및 워크플로우를 자동으로 검색합니다.   
따라서, 위 샘플들을 DevUI에서 바로 실행할 수 있습니다.

### 💻 실행 방법

터미널에서 다음의 명령을 실행하여 DevUI를 시작합니다.

> **⚠️ 주의**  MAF/DevUI 디렉토리에서 실행해야 합니다.

```bash
    cd .\MAF\DevUI
    devui . --port 8080
``` 

그러면, 현재 디렉토리 즉 DevUI 디렉토리에 존재하는 모든 에이전트와 워크플로우가 DevUI에 로드됩니다.

🌐 웹 브라우저에서 `http://localhost:8080`에 접속하여 DevUI 인터페이스를 확인할 수 있습니다.

> **💡 Tip**  
> 만일, 8080 포트가 이미 사용 중이면 올바로 구동이 되지 않을 수 있습니다.  
> 이 경우, 다른 포트 번호(예: 8090)를 지정하여 실행합니다.
